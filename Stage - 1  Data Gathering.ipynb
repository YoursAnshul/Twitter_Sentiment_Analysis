Consumer key is the API key that a service provider (Twitter, Facebook, etc.) issues to a consumer (a service that wants to access a user's resources on the service provider). This key is what identifies the consumer.

Consumer secret is the consumer "password" that is used, along with the consumer key, to request access (i.e. authorization) to a user's resources from a service provider.

Access token is what is issued to the consumer by the service provider once the consumer completes authorization.

```python
query='news'
max_tweets = 800
searched_tweets=[status for  status in tweepy.Cursor(api.search_tweets, q=query,lang="en").items(max_tweets)]
```

```python
searched_tweets[50]
```

The try block lets you test a block of code for errors. The except block lets you handle the error. 

```python
columns = set()
allowed_types = [str,int]
tweets_data = []

for status in searched_tweets:
    status_dict = dict(vars(status))
    keys = status_dict.keys()
    single_tweet_data = {}
    for k in keys:
        try:
            v_type = type(status_dict[k])
        except:
            v_type = None
            
        if v_type != None:
            
            if v_type in allowed_types:
                single_tweet_data[k] = status_dict[k]   
                columns.add(k)
    tweets_data.append(single_tweet_data)
        
```

```python
header_cols = list(columns)
```

```python
df = pd.DataFrame(tweets_data,columns=header_cols)
```
